```python
import numpy as np
import pandas as pd
import collections
from collections import Counter
```

```python
file1 = open('doc1.txt', 'r')
```

```python
file1contents = file1.read()
```

```python
file2 = open('doc2.txt', 'r')
file2contents = file2.read()

file3 = open('doc3.txt', 'r')
file3contents = file3.read()

file4 = open('doc4.txt', 'r')
file4contents = file4.read()
```

```python
file5 = open('doc5.txt', errors = 'ignore')
file5contents = file5.read()
#special characters in this file, needed to ignore so that we can read it
```

```python
file6 = open('doc6.txt', 'r')
file6contents = file6.read()
```

```python
all_files_contents = (file1contents + file2contents + file3contents + file4contents +file5contents + file6contents)
print(all_files_contents)
```

```python
all_files_contents.split()
```

```python
counter = Counter(all_files_contents.split())
```

```python
top_ten = counter.most_common(10)
```

```python
print(top_ten)
```

```python
print("| Word    | Frequency |")
for item in top_ten:
    print("|", item[0], " "*(6-len(item[0])), "|", 
          item[1], " "*(8-len(str(item[1]))), "|")
```
